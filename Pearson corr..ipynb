```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read data
df = pd.read_csv('prc.csv')
df
```

```python
# Transfor data type 
df['date'] = pd.to_datetime(df['date'])
```

# Pearson correlation

## January

```python
m1 = df[df['date'].dt.month == 1]
m1
```

```python
m1.groupby('state')[['confirmed','PM25']].corr(method="pearson")
```

```python
m1.groupby('state')[['confirmed','PM25']].corr(method="spearman")
```

```python
m1.groupby('state')[['confirmed','ozone']].corr()
```

```python
m1.groupby('state')[['confirmed','PM10']].corr()
```
